```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
a = 1
c = 0  # Example value for the constant c

# Parameters
m_P = 0.1
m_Q = 0
sigma_P = 0.2
sigma_Q = 0.1

# Calculating S
S = (m_Q * sigma_P**2 - m_P * sigma_Q**2) / (2 * sigma_P**2 * sigma_Q**2)

# Function to calculate k
def calculate_k(y):
    return np.exp((m_P**2 * sigma_Q**2 - m_Q**2 * sigma_P**2) / (2 * sigma_P**2 * sigma_Q**2)) * np.exp((sigma_Q**2 - sigma_P**2) / (2 * sigma_P**2 * sigma_Q**2) * (np.log(y))**2)

# Function z*(y)
def z_star(y):
    k = calculate_k(y)
    return -k / a * y**S + c

# Range of y values
y_values = np.linspace(0.1, 5, 500)

# Calculating z*(y) for each y
z_values = z_star(y_values)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(y_values, z_values, label=r'$z^*(y) = -\frac{k}{a} y^S + c$', color='blue')
plt.xlim(0, 5)  # Set x-axis limits
plt.ylim(-1.5, 0.5)  # Set y-axis limits
plt.xlabel('y')
plt.ylabel(r'$z^*(y)$')
#plt.title(r'Plot of $z^*(y)$')
plt.title('WITH VARIANCE SPREAD')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Let's look at some notation
# P: Close price
# C: Last price = call price
# k: strike price
#varrho: rate
#sigma: the standard deviation
# t: time of option signature
# T: option expiry time
# T_t = T-t
```

```python
#Let's use the "scipy.optimize.fsolve" function to find our sigma_Q


#TESLA

def y(T_t, k, P, varrho, sigma_Q):
    return (1 / (sigma_Q * np.sqrt(T_t))) * ((varrho - 0.5 * sigma_Q**2) * T_t - (k / P))


def call_price(T_t, k, P, varrho, sigma_Q):
    y0 = y(T_t, k, P, varrho, sigma_Q)
    return P * norm.cdf(y0 + sigma_Q * np.sqrt(T_t)) - k * np.exp(-varrho * T_t) * norm.cdf(y0)


def objective_function(sigma_Q, T_t, k, C, P, varrho):
    return call_price(T_t, k, P, varrho, sigma_Q) - C


def find_sigma_Q(T_t, k, C, P, varrho, initial_guess=0.2):
    sigma_Q = fsolve(objective_function, initial_guess, args=(T_t, k, C, P, varrho))
    return sigma_Q[0]




k_vals = [100, 105, 105, 100, 120, 110, 126.67, 125, 135, 110]
c_vals = [76.70, 72.20, 74.57, 76.65, 56.15,64.79, 53.10, 50.27, 39.89, 62.69]
p_vals = [176.29, 178.08, 178.79, 176.29, 176.29, 175.00, 176.29, 174.77, 175.00, 175.00]
ell_vals = [4,7,15,11,18,2,18,3,2,9]
varrho = 0.021853780098728594



sigma_vals=[]
for i in range(len(k_vals)):
    sigma_vals.append( find_sigma_Q(ell_vals[i]/252, k_vals[i], c_vals[i], p_vals[i], varrho) )

sigma1_Q = sum(sigma_vals)/len(sigma_vals)




print("The standard deviation obtained using TESLA is :", sigma1_Q)
```

```python
#Optimal allocation in the case of No spread variance
#constrainte q.q =1
m_P=np.array([m1_P,m2_P,m3_P,m4_P])
m_Q= np.array([varrho,varrho,varrho,varrho])

import numpy as np
from scipy.optimize import minimize

# Define the function psi(q)
def psi(q, C_P, C_Q, nu):
    C_P_q = np.dot(C_P, q)
    C_Q_q = np.dot(C_Q, q)
    term1 = np.dot(C_P_q, q)
    term2 = np.sqrt(2 * np.dot(C_P_q, q) * term1 - np.dot(C_Q_q, q)**2)
    term3 = np.dot(q, nu)
    return (term1 / term2) * np.exp((term3**2) / (2 * term1 - np.dot(C_Q_q, q)))

# Define the negative of psi(q) for minimization
def neg_psi(q, C_P, C_Q, nu):
    return -psi(q, C_P, C_Q, nu)

# Constraint: q.q = 1
def constraint(q):
    return np.dot(q, q) - 1

# Parameters (example values, you need to define them appropriately)

nu = m_P - m_Q # Example nu vector

# Initial guess for q
q0 = np.random.rand(4)
q0 /= np.linalg.norm(q0)  # Normalize to ensure q.q = 1

# Define constraints dictionary
constraints = {'type': 'eq', 'fun': constraint}

# Perform the optimization
result = minimize(neg_psi, q0, args=(C_P, C_Q, nu), constraints=constraints, method='SLSQP')

# Extract the optimal q
optimal_q = result.x
max_psi = -result.fun

print("Optimal q:", optimal_q)
print("Maximum psi(q):", max_psi)

# Verify the constraint
print("q.q:", np.dot(optimal_q, optimal_q))
```
